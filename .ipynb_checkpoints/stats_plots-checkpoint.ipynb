```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
numbers = [-5,1,2,3,3,3,4,4,5,6,7,7,12]
```

```python
fig, axs = plt.subplots(2,3)

# basic plot
axs[0, 0].boxplot(numbers)
axs[0, 0].set_title('basic plot with outliers')

# horizontal boxes
axs[0, 1].boxplot(numbers, 0, 'b', 0)
axs[0, 1].set_title('horizontal boxes with outliers')

# adjust spacing
fig.subplots_adjust(left=0.08, right=2, bottom=.25, top=2,
                    hspace=0.4, wspace=0.5)
```
